```python
import numpy as np 
import scipy as sp 
```

La funzione sp.linalg.lu fa una fattorizzazione con pivoting, ovvero fattorizzare una matrice $PA = LU$
- $P =$ matrice permutazione
- $A =$ matrice che vogliamo.
- $L =$ matrice inferiore
- $U =$ matrice superiore

Siccome che l'inversa di $P$ è uguale, possiamo scrivere anche come $A=PLU$
Se facciamo $\|A-PLU\|$ dovrebbe essere circa $0$.

```python
#problema test 1
A1=np.array([[9,3,6],[3,4,6],[0,8,8]])
#la funzione scipy.linalg.lu restituisce le tre matrici L,U,P tali che: B=P*L*LU
P,L,U=sp.linalg.lu(A1)
err=np.linalg.norm(A1-P@L@U)
print('norma errore LU ',err)

```

```python
#pb test 2
#la funzionescipy.linalg.lu_factor restituisce due matrici della fattorizzazione LU:
# # nella prima si memorizza  sia L(triangolare inferiore) che U (triangolare superiore)
# nella seconda si memorizza P 

A2=np.array([[3,-7,2,2],[-3,5,1,0],[6,-4,0,-5],[-9,5,-5,12]])
lu,piv=sp.linalg.lu_factor(A2)

print('lu',lu)
print('piv',piv)


x_true = np.ones((4,))

# b=np.array([-9,5,7,11])

# Costruisco il problema test con soluzione esatta x_true
b=A2@x_true

#la funzione scipy.linalg.lu_solve risolve un sistema lineare con il risultato della funzione 
#scipy.linalg.lu_factor

# La tupla lu,piv
x=sp.linalg.lu_solve((lu,piv),b)

print('sol=',x)
print(f"norma errore: {np.linalg.norm(x-x_true)}")
# print(np.linalg.norm(b-A2@x))
```

```python
#oppure si possono usare le funzioni della libreria numpy.linalg.solve che direttamente
# risolve il sistema lineare con una fattprizzazione LU con pivoting (che però rimane nascosta)


xx=np.linalg.solve(A2,b)

print('sol=',xx)
print(f"norma errore: {np.linalg.norm(xx-x_true)}")
#print(np.linalg.norm(b-A2@xx))
```
